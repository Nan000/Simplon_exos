```python
df.duplicated().any()
```

## Drop Outliers

```python
Q1 = df["Rent"].quantile(0.25)
Q3 = df["Rent"].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df["Rent"]<(Q1-1.5*IQR)) | (df["Rent"] > (Q3 + 1.5*IQR)))]
```

```python
sns.scatterplot(x = df["Size"] , y= df["Rent"], hue = df["City"]).set(title='Rent by Apartment Size and by Country');
```

```python
sns.lmplot(x="Size", y="Rent", data = df, hue = "City", scatter_kws = {'alpha':0.5},height = 6, aspect = 1.5).set(title='Rent by Apartment Size and by Country');
```

## Drop irrelevent data

```python
df.nunique()
```

```python
df = df.drop(["Posted On", "Floor", "Area Locality"], axis =1)
```

# Machine learning

```python
df.head(10)
```

```python
df_2 = df.drop(["Area Type"], axis =1)
df_3 = df.drop(["Area Type", "City"], axis =1)
```

```python
from sklearn.model_selection import train_test_split
y = df["Rent"]
X = df.drop(["Rent"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 6)
```

```python
y_2 = df_2["Rent"]
X_2 = df_2.drop(["Rent"], axis=1)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size= 0.3, random_state = 6)

y_3 = df_3["Rent"]
X_3 = df_3.drop(["Rent"], axis=1)
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(X_3, y_3, test_size= 0.3, random_state = 6)
```

## Numeric values Scaling

```python
from sklearn.preprocessing import MinMaxScaler
num_scaler = MinMaxScaler()
scaled_num_Xtrain = num_scaler.fit_transform(choose_numeric(X_train))
scaled_num_Xtest = num_scaler.transform(choose_numeric(X_test))
```

## Categorical values Scaling

```python
from sklearn.preprocessing import OneHotEncoder
cat_scaler = OneHotEncoder(handle_unknown='ignore')

cat_scaled_Xtrain = cat_scaler.fit_transform(choose_cat(X_train))
cat_scaled_Xtest = cat_scaler.transform(choose_cat(X_test))
```

## Concatenation

```python
scaled_num_Xtrain_df = pd.DataFrame(scaled_num_Xtrain).rename(columns = {0: "size"})
scaled_num_Xtest_df = pd.DataFrame(scaled_num_Xtest).rename(columns={0: "size"})

cat_scaled_Xtrain_df = pd.DataFrame.sparse.from_spmatrix(cat_scaled_Xtrain)
cat_scaled_Xtest_df = pd.DataFrame.sparse.from_spmatrix(cat_scaled_Xtest)

scaled_df_Xtrain = pd.concat([scaled_num_Xtrain_df,cat_scaled_Xtrain_df], axis =1)
scaled_df_Xtest = pd.concat([scaled_num_Xtest_df,cat_scaled_Xtest_df], axis = 1)
```

## Model : Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

model = lr.fit(scaled_df_Xtrain, y_train)
score = model.score(scaled_df_Xtest, y_test)
score
```

```python
score
```

```python
num_scaler_3 = MinMaxScaler()
scaled_num_Xtrain_3 = num_scaler_3.fit_transform(X_train_3)
scaled_num_Xtest_3 = num_scaler_3.transform(X_test_3)

lr_3 = LinearRegression()
model_3 = lr_3.fit(scaled_num_Xtrain_3, y_train_3)
score_3 = model_3.score(scaled_num_Xtest_3, y_test_3)
score_3
```

## Cross Validation

```python
from sklearn.model_selection import cross_validate

score_cv = cross_validate(model, scaled_df_Xtrain, y_train, cv=5)['test_score']
score_cv
```

```python
score_mean_cv = score_cv.mean()
score_mean_cv 
```

```python
f"LinearRegression score : {score} - Cross-Validation score : {score_cv} - CV_mean score : {score_mean_cv}"
```

## Prediction vs Observed

```python
import matplotlib.pyplot as plt

x = scaled_df_Xtest["size"]
y_predit = model.predict(scaled_df_Xtest)

plt.scatter(x, y_predit, c = "blue")
plt.scatter(x, y_test, c ="red")
plt.title("Rent predicted compared to Rent Observed by Scaled Apartment Size", fontdict=None, loc='center', pad=None)
plt.xlabel("Scaled Apartment Size")
plt.ylabel("Rent")
plt.show();
```

```python
fig, ax = plt.subplots()

x_tip = y_test
error = y_test - y_predit

ax.axhline(0, color='grey', linewidth=0.8)
ax.set_ylabel('Prediction Error')
ax.set_xlabel('Observed Rent')
ax.set_title('Prediction error by Rent')
plt.scatter(x_tip, error , c = "blue")

plt.show();
```

# Pipeline

```python
from sklearn import set_config
from sklearn.pipeline import make_pipeline

set_config(display='diagram')
```

## Pipeline Creation

```python
num_pipeline = make_pipeline(MinMaxScaler())
cat_pipeline = make_pipeline(OneHotEncoder(handle_unknown='ignore'))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector as selector

column_scaling = ColumnTransformer([
    ("num_transformation", num_pipeline, selector(dtype_exclude = "object")),
    ("cat_trasformation", cat_pipeline, selector(dtype_include = "object"))
])
```

```python
global_pipeline = make_pipeline(
    column_scaling,
    LinearRegression()
)
```

```python
column_scaling.fit(X_train, y_train)
```

```python
global_pipeline
```

## Apply Pipeline

```python
global_pipeline.fit(X_train, y_train)
```

```python
y_pred = global_pipeline.predict(X_test)
pred_error = y_pred - y_test
sns.histplot(x = pred_error, kde = True).set(title='Number of Prediction errors');
```

```python
global_pipeline.score(X_test, y_test)
```

```python
global_pipeline.fit(X, y)
score_1 = global_pipeline.score(X, y)
score_1
```

```python
import pickle

pickle.dump(global_pipeline, open("pipline.pkl", "wb"))
pickled_model = pickle.load(open("pipline.pkl", "rb"))
pickled_model.score(X,y)
```
